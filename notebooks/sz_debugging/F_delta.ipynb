# Equilibrium on the around the shear zone tooth

Given is $M_\mathrm{L}$ and $M_\mathrm{R}$

```python
import sympy as sp
sp.init_printing()
```

```python
F_Lc, F_Rc = \
sp.symbols(r'F_\mathrm{L}^\mathrm{c},F_\mathrm{R}^\mathrm{c}')
F_Ls, F_Rs = \
sp.symbols(r'F_\mathrm{L}^\mathrm{s},F_\mathrm{R}^\mathrm{s}')
delta_F_s, delta_F_c = \
sp.symbols(r'\Delta\!F^\mathrm{s}, \Delta\!F^\mathrm{s}')
```

```python
eq1 = sp.Eq(F_Lc, F_Rc + delta_F_c)
eq2 = sp.Eq(F_Ls, F_Rs + delta_F_s)
eq1, eq2
```

The cross sectional equilibrium in L and R cross section is expressed as 

$$
F_\mathrm{L}^\mathrm{c} = F_\mathrm{L}^\mathrm{s} = F_\mathrm{L}, \; 
F_\mathrm{R}^\mathrm{c} = F_\mathrm{R}^\mathrm{s} = F_\mathrm{R} \; 
$$

The normal force within the segment must satisfy the equilibrium conditions

$$
F_\mathrm{L}^\mathrm{c} = F_\mathrm{R}^\mathrm{c} + \Delta F^\mathrm{c} \\
F_\mathrm{L}^\mathrm{s} = F_\mathrm{R}^\mathrm{s} + \Delta F^\mathrm{s} 
$$
implying
$$
F_\mathrm{L} = F_\mathrm{R} + \Delta F 
$$

Then, the moment in the section R with the inner lever arm $d$ reads

$$
M_\mathrm{R} = F_\mathrm{R} d
$$

The section L can be now related to $\Delta F$ as

$$
M_\mathrm{L} = F_\mathrm{L} d = (F_\mathrm{R} + \Delta F)\, d
$$

```python
F_R, delta_F, d = sp.symbols(
'F_\mathrm{R}, \Delta\!F, d'
)
M_L = (F_R + delta_F) * d
M_L
```

Given a linear profile of the moment in the range $x \in (x_\mathrm{L},L)$  with the value zero at x = 0 we can express the moment $M_\mathrm{L}$ as 

$$
M_\mathrm{L} = 
M_\mathrm{R} + \frac{M_\mathrm{R}}{ L - x_\mathrm{R} } L_\mathrm{LR}
$$

```python
M_R, L, x_R, L_LR = sp.symbols(
    'M_\mathrm{R}, L, x_\mathrm{R}, L_\mathrm{LR}'
)
```

```python
M_L_ = M_R + M_R / (L-x_R) * L_LR
M_L_
```

```python
eq_M_L_ = sp.Eq(M_L, M_L_.subs(M_R, F_R*d))
eq_M_L_
```

```python
delta_F_solved = sp.solve( eq_M_L_, delta_F)[0]
delta_F_solved
```

The calculation using the assumption of a linear profile of the reinforcement force delivers $\Delta F$ which is relatively large.
