# Equilibrium on the around the shear zone teeth

$$
F_\mathrm{TR} = -F_\mathrm{CR}
$$

$$
F_\mathrm{CL} = F_\mathrm{CR} + \Delta F_\mathrm{C}
$$
$$
F_\mathrm{TL} = F_\mathrm{TR} + \Delta F_\mathrm{T}
$$

Since the normal force must be zero also on the right hand side, we obtain
$$
F_\mathrm{TL} = -F_\mathrm{CL} \implies F_\mathrm{CR} + \Delta F_\mathrm{C} = -F_\mathrm{TR} - \Delta F_\mathrm{T} \implies \Delta F_\mathrm{T} = \Delta F_\mathrm{C}
$$

The bending moment related to the reinforcement position
$$
 M_\mathrm{R} - M_\mathrm{L} - \Delta F_\mathrm{C} H = 0
$$

Since the moment profile is linear with the slope $Q$ and the distance between R-L is $L_\mathrm{cs}$, we can write

$$
M_\mathrm{R} = M_\mathrm{L} - Q L_\mathrm{cs}
$$

$$
\Delta F_\mathrm{C} = Q L_\mathrm{cs} / H
$$

$$
\Delta F_\mathrm{T} = Q L_\mathrm{cs} / H
$$

# Resolved case

$$
M_\mathrm{R} = F_\mathrm{RC}^{(+)} y^{(+)}_\mathrm{C} +  
F_\mathrm{RC}^{(-)} y^{(-)}_\mathrm{C} + F_\mathrm{RT} y_T
$$

$$
M_\mathrm{L} = M_\mathrm{R} + Q L_\mathrm{cs}
$$
