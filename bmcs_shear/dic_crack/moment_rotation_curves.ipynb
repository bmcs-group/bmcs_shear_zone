```python
n_C = len(dcl.items)
M_phi_C_ = []
for C in range(n_C):
    print('crack', C)
    cor = dcl.items[str(C)].cor
    M_phi_list = []
    #cor.dic_grid.T_stepping == 'delta_T'
    for t in cor.dic_grid.t_dic_T: #cor.dic_grid.tstring_time_F_T[1]
        cor.dic_grid.t = t
        cor.dic_grid.T_stepping == 'delta_T'
        cor.frame_position = 'inclined'
        print('t', t)
        M_phi_list.append([cor.M_t, cor.phi_t])
        #print(M_phi_list)
    M_phi = np.array(M_phi_list)
    M_phi_C_.append(M_phi)
M_phi_C = np.array(M_phi_C_)
```
